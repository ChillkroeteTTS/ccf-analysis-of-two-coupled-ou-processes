# Tau vs. E Analysis
This notebook visualizes a simulation where the parameters $\tau$ and $\epsilon$ are increased seperately.
An ensemble of the ccf's between the source and the delayed OU process is analyzed by peak height and width.

## Used Params

```python
import pandas as pd
def flatten(param):
    newP = dict(param)
    for k, v in newP['noiseType'].items():
        newP[k] = v
    
    newP.pop('noiseType', None)
    return newP

params = [flatten(res['p']) for res in results]
pd.DataFrame(params)
```

# 

```python
from plotting.plotting import gen_3d_data
import numpy as np
z_white = gen_3d_data([res for res in results if res['p']['noiseType']['type'] == NoiseType.WHITE], 'tau1', 'e')
z_red = gen_3d_data([res for res in results if res['p']['noiseType']['type'] == NoiseType.RED], 'tau1', 'e')

fig, axs = plt.subplots(1, 2)
plt.sca(axs[0])
axs[0].set_title('white noise')
plt.pcolormesh(np.arange(0, 1.2, 0.1), np.arange(0, 1.2, 0.1), z_white, vmin=0, vmax=1, cmap='OrRd')

plt.sca(axs[1])
axs[1].set_title('red noise')
plt.pcolormesh(np.arange(0, 1.2, 0.1), np.arange(0, 1.2, 0.1), z_red, vmin=0, vmax=1, cmap='OrRd')
plt.colorbar()

for ax in axs:
    ax.set_xlabel(r'$\tau_1$')
    ax.set_ylabel(r'$\epsilon$')
```
